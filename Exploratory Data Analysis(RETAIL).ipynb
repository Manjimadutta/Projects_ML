### Step-16: DETECTING OUTLIERS IN SOME FEATURES USING BOX-PLOT VISUALIZATION

```python
sns.boxplot(x=md1["Quantity"])
```

A lot of outliers from these four features have been observed. They may produce errors or false predictions if we don't remove them immediately. So, we now try to remove them from the dataset using interquartile range method.

### Step-17: APPLYING IQR METHOD TO REMOVE OUTLIERS PRESENT IN THE DATASET.

```python
Q1 = md1.quantile(0.25)
Q3 = md1.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

### Step-18: RESHAPING DATASET AFTER TAKING ALL THE OUTLIERS OUT

```python
md2 = md1[~((md1 < (Q1 - 1.5 * IQR)) |(md1 > (Q3 + 1.5 * IQR))).any(axis=1)]
print('The modified shape of the dataset is: ')
print(md2.shape)
```
