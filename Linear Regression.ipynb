[ IMPORTING REQUIRED LIBRARIES ]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import logging
print("Libraries imported successfully")
```

[ INSERTING THE DATA FROM THE GIVEN SOURCE ]

```python
url = "http://bit.ly/w-data"
md = pd.read_csv(url)
print("Data imported successfully")

md.head(10)
```

[ PLOTTING THE DATA USING SCATTER DIAGRAM ]

```python
md.plot(x='Hours', y='Scores', style='o', color='r')
plt.title('Diagram showing pattern of Hours and Scores')
plt.xlabel('Study hours')
plt.ylabel('Score Percentages')
plt.show()
```

[ PLOTTING THE DATA USING LINE DIAGRAM ]

```python
plt.style.use('ggplot')
md.plot(kind='line')
plt.title('Line diagram showing pattern of Hours and Scores')
plt.xlabel('Study Hours')
plt.ylabel('Score Percentages')
plt.show()
```

[ MODELLING THE DATA ]

```python
X = md.iloc[:, :-1].values
y = md.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print("Training complete.")
```

[ PLOTTING FITTED LINE OVER THE DATA ]

```python
line = regressor.coef_*X+regressor.intercept_

plt.scatter(X, y)
plt.plot(X, line, color='blue');
plt.title('Fitted diagram of Hours and Scores')
plt.show()
```

```python
print(X_test)
y_pred = regressor.predict(X_test)
```

```python
#comparing the actual data with the predicted data 

df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

[ PREDICTION OF SCORE OF A STUDENT WITH THE GIVEN STUDY HOUR VALUE ]

```python
hours = 9.25    #Given
test=np.array([hours])
test=test.reshape(-1,1)
pred = regressor.predict(test)
print("Hours studied per day = {}".format(hours))
print("Predicted Score = {}".format(pred[0]))
```
